### Log-Odds
* 개념

> 선형 회귀에서는 각 속성의 값에다가 계수(coefficient)에 각 곱하고 절편(intercept)을 더해서 예측 값(-∞에서 + ∞)을 구함.

> 로지스틱 회귀에서는 마지막에 예측 값 대신 log-odds를 구함.
